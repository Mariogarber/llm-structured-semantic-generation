```python
import posixpath
import os

envoy_path = r"C:\Users\mario\OneDrive\Documentos\UPM\Master_Data\TFM_Project\data\raw\Kubernetes\all"

compose_files = []
founded = []

for i in range(300):
    compose_path = posixpath.join(envoy_path, f"q{i}", "labeled_code.yml")
    if os.path.exists(compose_path):
        compose_files.append(compose_path)
        founded.append(i)
    compose_path_alt = posixpath.join(envoy_path, f"q{i}", "labeled_code.yaml")
    if os.path.exists(compose_path_alt):
        compose_files.append(compose_path_alt)
        founded.append(i)

print(f"Total docker-compose files found: {len(compose_files)}")
print(f"Total number of projects: {len(os.listdir(envoy_path))}")
```

```python
for file in compose_files:
    to_move_path = file.replace("raw", "docker_compose").replace("all", "")
    os.makedirs(os.path.dirname(to_move_path), exist_ok=True)
    # make a copy of the file
    with open(file, "r") as f:
        content = f.read()
    with open(to_move_path, "w") as f:
        f.write(content)
```

```python
for file in compose_files:
    question_file = file.replace("labeled_code.yaml", "question.txt").replace("labeled_code.yml", "question.txt")
    to_move_path = file.replace("raw", "question").replace("all", "").replace("labeled_code.yaml", "question.txt").replace("labeled_code.yml", "question.txt")
    os.makedirs(os.path.dirname(to_move_path), exist_ok=True)
    # make a copy of the file
    with open(question_file, "r") as f:
        content = f.read()
    with open(to_move_path, "w") as f:
        f.write(content)
```

```python
import os

INDEX = 1

kuber_root_path = r"C:\Users\mario\OneDrive\Documentos\UPM\Master_Data\TFM_Project\data\raw\Kubernetes"

folders_on_kuber = os.listdir(kuber_root_path)
folders_on_kuber
```

```python
# folders_on_kuber = folders_on_kuber[:4]

INDEX = 11

for folder in folders_on_kuber:
    if folder == "all":
        continue
    project_path = os.path.join(kuber_root_path, folder)
    print(f"Processing project: {project_path}")
    for q_folder in os.listdir(project_path):
        INDEX += 1
        print(f"Processing file: {q_folder}")
        # move all the content of the q_folder to a new folder with the name of the project and the index
        q_folder_path = os.path.join(project_path, q_folder)
        new_folder_path = os.path.join(kuber_root_path, "all", f"q{INDEX}")
        os.makedirs(new_folder_path, exist_ok=True)
        for file in os.listdir(q_folder_path):
            file_path = os.path.join(q_folder_path, file)
            new_file_path = os.path.join(new_folder_path, file)
            os.rename(file_path, new_file_path)
```
